```python
import os
import pandas as pd
```

```python
CLASSES = [
    "Alcantarillado", "Marca vial", "Hueco", "Parcheo", "Grietas longitudinales",
    "Grietas en forma de piel de cocodrilo", "Grietas transversales",
    "Meteorización y desprendimiento", "Espiras Magnéticas"
]

def classes2hot(list_classes):
    res = []
    for clase in CLASSES:
        res.append(1) if clase in list_classes else res.append(0)
    return res
```

```python
rootdir = 'classification_preds/by_target'

info = {"file_path":[], "num_classes":[], "truth":[], "predicted": []}

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        file_path = os.path.join(subdir, file)
        info["file_path"].append(file_path)
        info["num_classes"].append(int(subdir.split("/")[-2][:subdir.split("/")[-2].find("_")]))
        info["truth"].append(classes2hot([x.strip() for x in file_path[file_path.find("TARGET")+8:file_path.find("PREDICTED")-2].split("-")]))
        info["predicted"].append(classes2hot([x.strip() for x in file_path[file_path.find("PREDICTED")+11:file_path.find("__VIRB")].split("-")]))
```

```python
df_truth = pd.DataFrame(info["truth"], columns=CLASSES, index=info["file_path"])
df_truth.reset_index(inplace=True)
df_truth.rename(columns = {'index':'file_path'}, inplace=True)
df_truth['sum'] = df_truth.drop('file_path', axis=1).sum(axis=1)

df_predicted = pd.DataFrame(info["predicted"], columns=CLASSES, index=info["file_path"])
df_predicted.reset_index(inplace=True)
df_predicted.rename(columns = {'index':'file_path'}, inplace=True)
df_predicted['sum'] = df_predicted.drop('file_path', axis=1).sum(axis=1)
```
